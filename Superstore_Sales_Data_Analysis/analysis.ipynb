```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Sample - Superstore.csv', encoding="latin1")
df
```

```python
df.dropna(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])
```

```python
df["Order Year"] = df["Order Date"].dt.year
df["Order Month"] = df["Order Date"].dt.month
```

```python
df[["Sales", "Profit"]].sum()
```

```python
category_summary = df.groupby("Category")[["Sales", "Profit"]].sum().reset_index()
category_summary
```

```python
plt.figure()
sns.barplot(x="Category", y="Sales", data=category_summary)
plt.title("Sales by Category")
plt.show()
```

```python
sub_profit = df.groupby("Sub-Category")["Profit"].sum().sort_values(ascending=False).reset_index()
sub_profit
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x="Profit", y="Sub-Category", data=sub_profit)
plt.title("Profit by Sub-Category")
plt.show()
```

```python
region_sales = df.groupby("Region")["Sales"].sum().reset_index()
region_sales
```

```python
plt.figure()
plt.pie(region_sales["Sales"], labels=region_sales["Region"], autopct="%1.1f%%")
plt.title("Region-wise Sales Distribution")
plt.show()
```

```python
monthly_sales = df.groupby(["Order Year", "Order Month"])["Sales"].sum().reset_index()
monthly_sales
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x="Order Month", y="Sales", hue="Order Year", data=monthly_sales)
plt.title("Monthly Sales Trend")
plt.show()
```

```python
segment_summary = df.groupby("Segment")[["Sales", "Profit"]].sum().reset_index()
segment_summary
```

```python
plt.figure()
sns.barplot(x="Segment", y="Profit", data=segment_summary)
plt.title("Profit by Customer Segment")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x="Sales", y="Profit", data=df)
plt.title("Sales vs Profit")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df["Profit"], bins=50, kde=True)
plt.title("Profit Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x="Discount", y="Profit", data=df)
plt.title("Discount vs Profit")
plt.show()
```

```python
ship_sales = df.groupby("Ship Mode")["Sales"].sum().reset_index()

plt.figure(figsize=(6,4))
sns.barplot(x="Ship Mode", y="Sales", data=ship_sales)
plt.title("Sales by Ship Mode")
plt.xticks(rotation=30)
plt.show()
```

```python
pivot_table = pd.pivot_table(
    df,
    values="Sales",
    index="Category",
    columns="Segment",
    aggfunc="sum"
)

plt.figure(figsize=(6,4))
sns.heatmap(pivot_table, annot=True, fmt=".0f", cmap="coolwarm")
plt.title("Sales Heatmap: Category vs Segment")
plt.show()
```

```python
yearly_sales = df.groupby("Order Year")["Sales"].sum().reset_index()

plt.figure(figsize=(6,4))
sns.lineplot(x="Order Year", y="Sales", data=yearly_sales, marker="o")
plt.title("Yearly Sales Trend")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x="Quantity", y="Profit", data=df)
plt.title("Quantity vs Profit")
plt.show()
```

```python
category_qty = df.groupby("Category")["Quantity"].sum().reset_index()

plt.figure(figsize=(6,4))
sns.barplot(x="Category", y="Quantity", data=category_qty)
plt.title("Total Quantity Sold by Category")
plt.show()
```

### All Over Observations :
## 📌 Business Summary (Key Insights)

- Sales growth does not always lead to higher profit due to heavy discounting.

- Technology category is the strongest performer in both sales and profitability.

- Furniture category shows lower profit margins and needs cost optimization.

- High discounts are the primary reason for loss-making orders.

- Consumer segment contributes the maximum revenue across all categories.

- West and East regions generate the highest share of total sales.

- Sales show seasonal trends with noticeable peaks during certain months.

- A small number of states contribute a large portion of overall revenue.

- Standard shipping mode is preferred by customers due to lower cost.
